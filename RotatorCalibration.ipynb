```python
"""Polarization rotator calibration tool.
   
This tool takes the output from a cockpit Rotator Sweep experiment
and produces transmitted instensity profiles as a function of
rotator voltage.
    
:copyright: 2015-2016 by Mick Phillips & the University of Oxford.
:license: GPL. See LICENSE-GPL for more details
"""

# Imports
from PyQt4 import QtCore, QtGui
import matplotlib.pyplot as plt
import os.path
import sys
import Mrc
import numpy as np
import re
plt.rcParams['image.cmap'] = 'gray'
app=QtGui.QApplication.instance()
if not app:
    app = QtGui.QApplication(sys.argv)
if not globals().get('path'):
    path = '.'

# Match string for extracting voltage sweep parameters.
rotatorParamRe = r"Rotator start and delta: \[(?P<start>[-+]?(\d+(\.\d*)?|\.\d+)([eE][-+]?\d+)?), (?P<delta>[-+]?(\d+(\.\d*)?|\.\d+)([eE][-+]?\d+)?)\]"
    
class Fitter():
    """Fits a polynomial between points defined by clicks."""
    def __init__(self, axis, data):
        # X and Y data
        self.data = data
        # Axis on the plot.
        self.axis = axis
        # Limits for the fit.
        self.limits = [None, None]
        # Keep track of first few clicks.
        self.clickCount = 0
        # Fit line.
        self.line = axis.plot(data[0][0], data[1][0], 'm-', lw=4, zorder=len(axis.lines))[0]
        # Fit line label.
        self.label = axis.annotate('', (0, 0), ha='center',
                                   arrowprops=dict(facecolor='magenta',
                                                   width=3, shrink=0.1))
        # MPL connection ID.
        self.cid = axis.figure.canvas.mpl_connect('button_press_event', self)       
        
        
    def __call__(self, event):
        """Update fit limits and do the fit."""
        if self.clickCount == 0:
            self.limits[0] = event.xdata
            self.clickCount += 1
        elif self.clickCount == 1:
            self.limits[1] = event.xdata
            self.clickCount += 1
        else:
            which = np.abs([lim - event.xdata for lim in self.limits]).argmin()
            self.limits[which] = event.xdata
        if None not in self.limits:
            self.doFit()

            
    def doFit(self):
        """Do the fit and update the plot."""
        # Extract data subset for fitting.
        condition = np.logical_and(self.data[0] >= min(self.limits), self.data[0] <= max(self.limits))
        xData = np.extract(condition, self.data[0])
        yData = np.extract(condition, self.data[1])
        # Do the fit.
        fitParams = np.polyfit(xData, yData, 2)
        # Update the fit line on the plot.
        xFit = np.linspace(min(self.limits), max(self.limits), 64)
        yFit = np.polyval(fitParams, xFit)     
        self.line.set_data(xFit, yFit)
        # Determine peak position.
        xAtVertex = -fitParams[1] / (2. * fitParams[0]) # from dy/dx = 0
        yAtVertex = np.polyval(fitParams, xAtVertex)
        # Update the label text.
        self.label.set_text('%.3f' % xAtVertex)
        # Update the label position
        yMidPoint = np.mean(self.axis.get_ylim())
        if yAtVertex < yMidPoint:
            labelY = min(self.axis.get_ylim()) + 0.1 * np.subtract(*self.axis.get_ylim())
        else:
            labelY = max(self.axis.get_ylim()) - 0.05 * np.subtract(*self.axis.get_ylim())       
        self.label.xytext = ((xAtVertex, labelY))
        self.label.xy = (xAtVertex, yAtVertex)
        # Redraw the canvas.
        self.axis.figure.canvas.draw()

        
def makeProfile(dataPath):
    """Make profile of counts vs. voltages from image stack at dataPath."""
    # Load the data.
    data = Mrc.bindFile(dataPath)
    # Size and shape variables.
    nV, ny, nx = data.shape
    # Extract sweep parameters from the header.
    vStart = None
    vDelta = None
    for t in data.Mrc.hdr.title:
        match = re.match(rotatorParamRe, t)
        if match:
            vStart = float(match.groupdict()['start'])
            vDelta = float(match.groupdict()['delta'])
    if None in (vStart, vDelta):
        raise Exception('Could not find sweep parameters in header.')
    vEnd = vStart + nV * vDelta
    # Array of voltages.
    voltages = np.arange(vStart, vEnd, vDelta)
    # Sum counts in each image in the stack.
    counts = np.sum(np.sum(data, axis=2), axis=1)
    return voltages, counts
    
```

```python
# Ask for data paths.
paraPath = str(QtGui.QFileDialog.getOpenFileName(None, "Select parallel polarization data file...", path))
path = os.path.dirname(paraPath) # Update path so next dialog opens in same folder.
perpPath = str(QtGui.QFileDialog.getOpenFileName(None, "Select perpendicular polarization data file...", path))

# Make profiles for each dataset.
paraVoltages, paraCounts = makeProfile(paraPath)
perpVoltages, perpCounts = makeProfile(perpPath)

# Ratio the data over the range of common voltages.
vMin = max(paraVoltages.min(), perpVoltages.min())
vMax = min(paraVoltages.max(), perpVoltages.max())
vNum = min(np.count_nonzero(np.logical_and(paraVoltages >= vMin, paraVoltages <= vMax)),
           np.count_nonzero(np.logical_and(perpVoltages >= vMin, perpVoltages <= vMax)))
voltages = np.linspace(vMin, vMax, vNum)
# Calculate the ratio, using interpolation to match voltages.
ratio = np.interp(voltages, paraVoltages, paraCounts) / np.interp(voltages, perpVoltages, perpCounts)

# Figure and left-hand axis.
fig, axl = plt.subplots()
# Right-hand axis.
axr = axl.twinx()
# Plot profiles using the left-hand axis.
axl.plot(perpVoltages, perpCounts, 'r-')
axl.plot(paraVoltages, paraCounts, 'b-')
# Plot ratio using the right-hand axis.
axr.plot(voltages, ratio, 'k-')
fitter = Fitter(axr, (voltages, ratio))
```
