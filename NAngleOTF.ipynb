```python
"""NAngleOTF

This tool evaluates the structured illumination optical transfer function 
from a beadstack with N illumination orientations.

:copyright: 2016 by Mick Phillips & the University of Oxford.
:license: GPL. See LICENSE-GPL for more details
"""

# Imports
from PyQt4 import QtCore, QtGui
import matplotlib.pyplot as plt
import os.path
import sys
import Mrc
import numpy as np

# Global settings
plt.rcParams['image.cmap'] = 'gray'
app=QtGui.QApplication.instance()
if not app:
    app = QtGui.QApplication(sys.argv)
if not globals().get('path'):
    path = '.'
```

```python
# This cell sets dataset-specific settings and allows data selection.
nPhases = 5
nOrders = 3
nAngles = 3
gamma = 0.15
swDataPath = str(QtGui.QFileDialog.getOpenFileName(None, "Select data file...", path))
path = os.path.dirname(swDataPath)

# Load data
data = Mrc.bindFile(swDataPath)
```

```python
# This cell does the calculation.
def makeMatrix(nPhases):
    """Return a separation matrix for data with nPhases."""
    nOrders = (nPhases+1)/2
    sepMatrix=np.zeros((nPhases,nPhases), dtype=np.float)
    phi = 2 * np.pi / nPhases
    for phase in xrange(nPhases):
        sepMatrix[0][phase] = 1.0/nPhases
        for order in range (1, nOrders):
            sepMatrix[2*order-1][phase]=2.0*np.cos(phase*order*phi)/nPhases
            sepMatrix[2*order][phase]=2.0*np.sin(phase*order*phi)/nPhases
    return sepMatrix


def findPeak(phaseStack):
    """Return the position of the bead centre.

    The centre position is estimated by fitting a prabola to pseudo-widefield data.
    """
    # Make pseudo-widfield stack.
    nPhases, nz, ny, nx = phaseStack.shape
    summedPhases = np.zeros((nz, ny, nx))
    for z in range(nz): summedPhases[z,:,:] = phaseStack[:,z,:,:].sum(axis=0) / nPhases
    peakZ, peakY, peakX = np.unravel_index(summedPhases.argmax(), (nz, ny, nx))
    # 2nd order polynomial fits to peak.
    xFit = np.polyfit(xrange(peakX-2, peakX+3),
                      summedPhases[peakZ, peakY, peakX-2:peakX+3],
                      2)
    yFit = np.polyfit(xrange(peakY-2, peakY+3), 
                      summedPhases[peakZ, peakY-2:peakY+3, peakX],
                      2)
    zFit = np.polyfit(xrange(peakZ-2, peakZ+3),
                      summedPhases[peakZ-2:peakZ+3, peakY,peakX],
                      2)
    peakFit = (-(zFit[1]/(2*zFit[0])),
               -(yFit[1]/(2*yFit[0])),
               -(xFit[1]/(2*xFit[0])))
    return np.array(peakFit)


def separate(phaseStack):
    """Apply the sepration matrix to an phase/z/y/x stack."""
    nPhases, nz, ny, nx = phaseStack.shape
    bands = np.zeros(phaseStack.shape)
    sepMatrix = makeMatrix(nPhases)
    for j in range(nPhases):
        for i in range(nPhases):
            bands[i,...] += phaseStack[j,...] * sepMatrix[i][j]
    return bands


def combineOTF(otf):
    """Combine the real and imaginary parts of the OTF."""
    nPhases, nz, ny, nx = otf.shape
    nOrders = (nPhases+1)/2
    combined = np.zeros((nOrders, nz, ny, nx), dtype=complex)
    # 0-order is unchanged.
    combined[0,...] = otf[0,...]
    for order in xrange(1, nOrders):
        # Determine the phase shift relative to zero-order.
        bandReMag = otf[2*order-1].sum()
        bandImMag = otf[2*order].sum()  
        phi = np.arctan(bandImMag/bandReMag)
        # This gives phi is in the first quadrant. Work out which quadrant
        # it should lie in from kz=0 plane.
        if (otf[2*order-1].flat[nx/4*nz].real < 0 and otf[2*order].flat[nx/4*nz].real >0):
            phi += np.pi;
        elif (otf[2*order-1].flat[nx/4*nz].real < 0 and otf[2*order].flat[nx/4*nz].real <0):
            phi = np.pi-phi;
        elif (otf[2*order-1].flat[nx/4*nz].real > 0 and otf[2*order].flat[nx/4*nz].real >0):
            phi = -phi;
        combined[order,...].real = otf[2*order-1,...].real * np.cos(phi).real + otf[2*order,...].real * np.sin(phi).real
        combined[order,...].imag = otf[2*order-1,...].imag * np.cos(phi).real + otf[2*order,...].imag * np.sin(phi).real
    return combined


# Size and shape variables.
nz, ny, nx = data.shape
nz /= nPhases * nAngles
nzyx = np.array((nz, ny, nx))

dr = data.Mrc.hdr.m[1]
dz = data.Mrc.hdr.m[2]

dkr = 1./(ny*dr);
dkz = 1./(nz*dz);

# 2D Hanning window for apodization.
window = np.add(np.hanning(ny)[None,...], np.hanning(nx)[...,None]) / 2


results = []

for angle in range(nAngles):
    angleData = data[angle*nPhases*nz:(angle+1)*nPhases*nz]
    # Stack with indices [phase, z, y ,x]. Apodization happens here.
    phaseStack = np.zeros((nPhases, nz, ny, nx))
    for phase in xrange(nPhases):
        phaseStack[phase,0:nz,...] = window * angleData[phase:nz*nPhases:nPhases,...]

    # Estimate the bead position.
    peakPos = findPeak(phaseStack)

    # Background?

    # Separate the bands.
    bands = separate(phaseStack)

    # OTF from phase-corrected FFT.
    # Bead-offset phase error correction.
    dPhiZ = 2*np.pi*peakPos[0]/nz
    dPhiY = 2*np.pi*peakPos[1]/ny
    dPhiX = 2*np.pi*peakPos[2]/nx
    kzRange = np.fft.ifftshift(np.arange(-nz/2+nz%2, nz/2+nz%2))
    kyRange = np.fft.ifftshift(np.arange(-ny/2+ny%2, ny/2+ny%2))
    kxRange = np.fft.ifftshift(np.arange(-nx/2+nx%2, nx/2+nx%2))
    kz, ky, kx = np.meshgrid(kzRange, kyRange, kxRange, indexing='ij')
    eiphi = np.exp(np.complex(0, 1)*(kz*dPhiZ + ky*dPhiY + kx*dPhiX))
    otf = np.zeros(bands.shape, dtype=complex)
    for band in xrange(nPhases):
        otf[band] = eiphi * np.fft.fftn(window * bands[band])

    # Bead size compensation?

    # 'Modify'?

    # 'Cleanup'?

    # 'FixOrigin'?

    # 'Rescale'?

    # Combine real and imaginary OTF components.
    combined = combineOTF(otf)

    # Radially average about kz=0.
    ky, kx = np.ogrid[-ny/2:ny/2, -nx/2:nx/2]
    kz = np.ogrid[0:nz]
    # Radius in k-space, shifted to match FFT.
    radius = np.fft.fftshift(np.hypot(kx, ky))
    # Assign each pixel to a bin in r_circ.
    r_circ = np.arange(int(np.max(radius)))
    circles = np.digitize(radius.flat, r_circ)
    # Sum around circles.
    s = np.array([[[combined[order,k,...].flat[circles==r].sum(axis=-1) for r in r_circ] for k in kz] for order in xrange((nPhases+1)/2)])
    avg = np.divide(s, np.bincount(circles)[1:])

    results.append(avg)
    
    
# Plot the results.
for angle, avg in enumerate(results):
    kxLim = 2 * avg.shape[-1]/3
    fig, axes = plt.subplots(2, 3)
    fig.suptitle('Angle %d' % angle)
    for band, axis in enumerate(axes[0]):
        #axis.imshow(np.flipud(np.log(np.abs(avg[band].T))))
        axis.imshow(np.flipud(np.power(np.abs(avg[band].T),gamma) / (np.power(255,gamma))))
        axis.set_xticks([])
        axis.set_yticks([])
        if band==0:
            axis.set_ylabel('abs')
    for band, axis in enumerate(axes[1]):
        axis.imshow(np.flipud(np.angle(avg[band].T)))
        axis.set_xticks([])
        axis.set_yticks([])
        if band==0:
            axis.set_ylabel('phase')
            axis.set_xlabel('0')
        else:
            axis.set_xlabel('+/-%d' % band)
    fig.subplots_adjust(bottom=0.05, left=0.05, top=0.95, right=0.95)
    fig.show()
```
